# Notebook on scraping Finn.no advertisments 

```python
import requests
from bs4 import BeautifulSoup
import csv
import bs4, re
import pandas as pd 
```

```python
base_url = "https://www.finn.no/car/used/search.html?model=1.818.2783&page="
ending = "&sort=PUBLISHED_DESC"
```

```python
id_number = []

for i in range (1, 6, 1):
    url = base_url + str(i) + ending 
    source = requests.get(url).text
    soup = BeautifulSoup(source, "lxml")

    for article in soup.find_all("article"):
        id_ = article.find("a")
        id_number.append(id_["id"]) 
```

```python
def transform(soup):
    article = soup.find("div", class_= "grid__unit u-r-size2of3")
        
    headline = article.find("h1", class_ = "u-t2 u-word-break").text
        
    media_grid = article.find_all("div", class_="u-strong")
        
    year = media_grid[0].get_text()
        
    transmission = media_grid[2].get_text()
        
    km = media_grid[1].get_text().replace(" km", "")
        
    total_price = article.find("span", class_="u-t3").text.replace("kr", "")
        
    specifications = article.find("dl", class_="list-descriptive u-col-count2 u-col-count3from990")
        
    power = specifications.find(text=re.compile('Hk')).replace(" Hk", "")
        
    try:
        cylinder_capacity = specifications.find(text=re.compile(" l")).replace(" l", "")
    
    except:
        cylinder_capacity = "NaN"
        
    try:
         CO2_emission = specifications.find(text=re.compile(" g/km")).replace(" g/km", "")
        
    except:
        CO2_emission = "NaN"
    
    
    car = {
    "headline": headline,
    "year" : year,
    "transmission" : transmission,
    "km" : km,
    "total_price" : total_price,
    "power" : power,
    "cylinder_capacity" : cylinder_capacity,
    "CO2_emission" : CO2_emission
    }
    
    data.append(car)
```

```python
for x in id_number:
    c = extract(x)
    transform(c)
```
